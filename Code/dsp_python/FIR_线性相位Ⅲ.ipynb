# FIR滤波器结构：第Ⅲ类线性相位型

画出单位脉冲响应为 h(n)=0.5,-0.5,1,-1,0,1,-1,0.5,-0.5 的第III型线性相位滤波器的幅度函数和相位函数。

```python
#library
import numpy as np;from math import *
import matplotlib.pyplot as plt;from matplotlib import ticker

#单位脉冲响应
h = np.array([0.5,-0.5,1,-1,0,1,-1,0.5,-0.5]) #N为奇数，奇对称
N = len(h);L = int((N-1)/2)

#幅度函数和相位函数
c = 2*h[L::-1] #c(n)各值
n = np.arange(L+1);w = np.arange(501)*2*pi/500
c = c.reshape(-1,1);w = w.reshape(-1,1) #将c(n)和w(n)转为一列
Hw = np.dot(np.sin(w*n),c);Pw = -w*L+pi/2 ##幅度函数和相位函数

#绘制单位响应
fig,ax = plt.subplots();ax.stem(h,basefmt="")
ax.set_title('第III类线性相位的h(n)');ax.set_xlabel('n')
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来显示负号
fig.savefig('./fir_linear_phase3_1.png',dpi=500)

#绘制幅度函数和相位函数
fig,axs = plt.subplots(2,1,constrained_layout=True)
axs[0].plot(w/pi,Hw);axs[1].plot(w/pi,Pw/pi)
axs[0].set_title('第III类线性相位的幅度函数');axs[0].grid()
axs[0].set_xlabel('frequency');axs[0].set_ylabel(r'$ H( \omega )$')
formatter = ticker.FormatStrFormatter('%1.2f$ \pi$')
axs[0].xaxis.set_major_formatter(formatter)
axs[1].set_title('第III类线性相位的相位函数');axs[1].grid()
axs[1].set_xlabel('frequency');axs[1].set_ylabel(r'$ \theta ( \omega )$')
axs[1].xaxis.set_major_formatter(formatter)
axs[1].yaxis.set_major_formatter(formatter)
plt.show();fig.savefig('./fir_linear_phase3_2.png',dpi=500)
```
