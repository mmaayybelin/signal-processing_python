# 栅栏效应

```python
#导入使用的库
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt

f = 8000
N = 512#自己改N，相应长度也要改
t = np.arange(0, 1, 1/f)
x = np.sin(2*np.pi*100*t)  + 2*np.sin(2*np.pi*300*t)
A = x[:N]
y = np.fft.rfft(A,N)/(N)
freqs = np.linspace(0, int(f/2), int(N/2+1))
B = np.abs(y)
plt.figure()
plt.subplot(211)
plt.plot(t[:N], A)
plt.xlabel("t/s")
plt.subplot(212)
plt.plot(freqs, B)
plt.xlabel("f(Hz)")
plt.show()
```
