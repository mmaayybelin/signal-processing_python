# 混叠效应

```python
#导入使用的库
import numpy as np
import pylab as pl

fs = 900#自己改频率
N = 512
t = np.arange(0, 1, 1/fs)
x = np.sin(2*np.pi*100*t)+2*np.sin(2*np.pi*400*t)

xs = x[:N]
xf = np.fft.rfft(xs)/N
freqs = np.linspace(0, int(fs/2), int(N/2+1))
xfp = np.abs(xf)
pl.figure()
pl.subplot(211)
pl.plot(t[:N], xs)
pl.xlabel("Time(S)")
pl.subplot(212)
pl.plot(freqs, xfp)
pl.xlabel("Freq(Hz)")
pl.show()
```
